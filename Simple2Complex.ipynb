```python
4️⃣ E-commerce customer analysis
Sample dataset orders.csv
csv
order_id,customer_id,order_date,order_value,is_new_customer
1,101,2023-01-05,120.5,True
2,102,2023-01-06,80.0,True
3,101,2023-02-10,60.0,False
4,103,2023-02-15,200.0,True
Code
python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("orders.csv", parse_dates=["order_date"])

# Basic features
df["month"] = df["order_date"].dt.to_period("M")

# New vs repeat customers
customer_orders = df.groupby("customer_id")["order_id"].count()
df["customer_type"] = df["customer_id"].map(
    lambda cid: "New" if customer_orders[cid] == 1 else "Repeat"
)

# Average order value overall and by type
aov_overall = df["order_value"].mean()
aov_by_type = df.groupby("customer_type")["order_value"].mean()

# Monthly revenue
monthly_revenue = df.groupby("month")["order_value"].sum()

print("AOV overall:", aov_overall)
print("\nAOV by customer type:\n", aov_by_type)
print("\nMonthly revenue:\n", monthly_revenue)

sns.barplot(x=df["customer_type"], y=df["order_value"])
plt.title("Order Value by Customer Type")
plt.show()
5️⃣ Financial data analysis
Sample dataset prices.csv
csv
date,close
2023-01-02,100
2023-01-03,102
2023-01-04,101
2023-01-05,105
2023-01-06,103
Code
python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("prices.csv", parse_dates=["date"])
df = df.sort_values("date")

# Daily returns
df["return"] = df["close"].pct_change()

# Moving averages
df["ma_3"] = df["close"].rolling(window=3).mean()
df["ma_5"] = df["close"].rolling(window=5).mean()

# Volatility (rolling std of returns)
df["vol_3"] = df["return"].rolling(window=3).std()

print(df)

plt.figure(figsize=(8,4))
plt.plot(df["date"], df["close"], label="Close")
plt.plot(df["date"], df["ma_3"], label="MA 3")
plt.plot(df["date"], df["ma_5"], label="MA 5")
plt.legend()
plt.title("Price with Moving Averages")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
6️⃣ Movie ratings analysis
Sample dataset movies.csv
csv
movie_id,title,genre,rating,votes,director,lead_actor
1,Movie A,Action,7.5,1200,Dir X,Actor P
2,Movie B,Drama,8.2,800,Dir Y,Actor Q
3,Movie C,Action,6.9,500,Dir X,Actor R
4,Movie D,Comedy,7.8,300,Dir Z,Actor P
Code
python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")

# Best genres by average rating
genre_rating = df.groupby("genre")["rating"].mean().sort_values(ascending=False)

# Ratings vs votes
print("Genre rating:\n", genre_rating)

sns.scatterplot(data=df, x="votes", y="rating", hue="genre")
plt.title("Ratings vs Votes")
plt.show()

# Director performance
director_rating = df.groupby("director")["rating"].mean().sort_values(ascending=False)
print("\nDirector performance:\n", director_rating)

# Actor performance
actor_rating = df.groupby("lead_actor")["rating"].mean().sort_values(ascending=False)
print("\nActor performance:\n", actor_rating)
7️⃣ Interactive data dashboard (Plotly + Dash)
Sample dataset kpi_sales.csv
csv
date,region,revenue,orders
2023-01-01,North,1000,10
2023-01-01,South,800,8
2023-01-02,North,1200,11
2023-01-02,South,900,9
Code (dashboard.py)
python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

df = pd.read_csv("kpi_sales.csv", parse_dates=["date"])

app = Dash(__name__)

app.layout = html.Div([
    html.H2("Sales Dashboard"),
    dcc.Dropdown(
        id="region-filter",
        options=[{"label": r, "value": r} for r in df["region"].unique()],
        value=None,
        placeholder="Select region (optional)",
        multi=True
    ),
    dcc.Graph(id="revenue-trend"),
    dcc.Graph(id="orders-trend")
])

@app.callback(
    [Output("revenue-trend", "figure"),
     Output("orders-trend", "figure")],
    [Input("region-filter", "value")]
)
def update_charts(selected_regions):
    dff = df.copy()
    if selected_regions:
        dff = dff[dff["region"].isin(selected_regions)]

    rev_fig = px.line(dff, x="date", y="revenue", color="region", title="Revenue Over Time")
    ord_fig = px.line(dff, x="date", y="orders", color="region", title="Orders Over Time")
    return rev_fig, ord_fig

if __name__ == "__main__":
    app.run_server(debug=True)
Run with:

bash
python dashboard.py
8️⃣ Web scraping + analysis
Example: scrape simple job titles (static HTML) and analyze word frequency
Note: This assumes a simple static page; for real sites you must respect robots.txt and TOS.

python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://example.com/jobs"  # placeholder
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Example: assume job titles are in <h2 class="job-title">
titles = [h2.get_text(strip=True) for h2 in soup.select("h2.job-title")]

df = pd.DataFrame({"job_title": titles})

# Simple word frequency
all_words = " ".join(df["job_title"]).lower().split()
word_series = pd.Series(all_words)
word_counts = word_series.value_counts().head(20)

print("Top words in job titles:\n", word_counts)
You can swap the selector and URL for product prices, etc.

9️⃣ Customer churn analysis (basic ML)
Sample dataset churn.csv
csv
customer_id,tenure,monthly_charges,total_charges,churn
1,5,70.5,352.5,Yes
2,24,55.0,1320.0,No
3,12,80.0,960.0,Yes
4,36,60.0,2160.0,No
Code
python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

df = pd.read_csv("churn.csv")

# Encode target
df["churn_flag"] = df["churn"].map({"Yes": 1, "No": 0})

features = ["tenure", "monthly_charges", "total_charges"]
X = df[features]
y = df["churn_flag"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print(classification_report(y_test, y_pred))
🔟 End-to-end data analytics project
Let’s treat this as a sales + customers mini end-to-end flow.

Sample dataset end_to_end_sales.csv
csv
order_id,order_date,customer_id,region,order_value
1,2023-01-01,101,North,120
2,2023-01-02,102,South,80
3,2023-01-05,101,North,60
4,2023-02-01,103,West,200
Code (cleaning → EDA → simple insights)
python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load
df = pd.read_csv("end_to_end_sales.csv", parse_dates=["order_date"])

# 2. Basic cleaning
df = df.dropna(subset=["order_value"])
df = df[df["order_value"] > 0]

# 3. Feature engineering
df["month"] = df["order_date"].dt.to_period("M")

# 4. KPIs
monthly_rev = df.groupby("month")["order_value"].sum()
region_rev = df.groupby("region")["order_value"].sum()
customer_ltv = df.groupby("customer_id")["order_value"].sum().sort_values(ascending=False)

print("Monthly revenue:\n", monthly_rev)
print("\nRevenue by region:\n", region_rev)
print("\nCustomer LTV:\n", customer_ltv)

# 5. Visuals
sns.barplot(x=region_rev.index.astype(str), y=region_rev.values)
plt.title("Revenue by Region")
plt.show()

monthly_rev.plot(kind="bar", title="Monthly Revenue")
plt.show()
```
