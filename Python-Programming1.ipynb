```python
✅ BATCH 5 — Exercises 13 to 15
⭐ EXERCISE 13 — Multiplication Table (1–10)
Question
Print the multiplication table from 1 to 10.

✔️ Canonical Solution
python
for i in range(1, 11):
    for j in range(1, 11):
        print(i * j, end=" ")
    print()
⭐ Alternative Solutions (from your document)
1. Clean nested loops
python
for i in range(1, 11):
    for j in range(1, 11):
        print((i * j), end=" ")
    print()
2. Pythonic list comprehension
python
[[print(x*y, end=" ") for x in range(1,11)] and print() for y in range(1,11)]
3. Function version
python
def MultiTable(n):
    for i in range(1, n+1):
        for j in range(1, 11):
            print(i*j, end=" ")
        print("")

MultiTable(10)
4. While‑loop version
python
i = 1
while i <= 10:
    j = 1
    while j <= 10:
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
📝 Notes
The list‑comprehension version is clever but less readable.

The canonical nested‑loop version is best for beginners.

⭐ EXERCISE 14 — Downward Star Pyramid
Question
Print the pattern:

Code
*****
****
***
**
*
✔️ Canonical Solution
python
for i in range(5, 0, -1):
    print("*" * i)
⭐ Alternative Solutions (from your document)
1. Using sep and end
python
for i in range(5, 0, -1):
    print('*' * i, sep=' ', end="\n")
2. Using a function
python
def star_pyramid(n):
    for i in range(n, 0, -1):
        print("*" * i)

star_pyramid(5)
3. Using while loop
python
i = 5
while i > 0:
    print("*" * i)
    i -= 1
📝 Notes
All solutions are correct; the simplest is best.

⭐ EXERCISE 15 — Exponent Function
Question
Write a function to compute:

Code
base^exponent
✔️ Canonical Solution
python
def exponent(base, exp):
    return base ** exp
⭐ Alternative Solutions (from your document)
1. Manual multiplication
python
base = 2
exponent = 3
result = 1

for i in range(exponent):
    result *= base

print(result)
2. Using pow()
python
def exponent(b, e):
    print(pow(b, e))

exponent(2, 5)
3. Expanded print format
python
base = 2
exp = 3
print(f"{base} raises to the power of {exp} is {base**exp}")
4. Step‑by‑step multiplication display
python
base = 2
exp = 3

print(base, "raises to the power of", exp, ":", base**exp, end=" ")
print("( ", end="")

for i in range(exp-1):
    print(base, "*", end=" ")

print(base, "=", base**exp, ")")
5. Input‑validated version
python
def exponent_check():
    while True:
        exponent = input("enter the exponent: ")
        if exponent.isdigit() and int(exponent) > 1:
            return int(exponent)

def integer_check():
    while True:
        base = input("enter a base: ")
        try:
            return int(base)
        except ValueError:
            pass

def expression(value, exponent):
    return value ** exponent

base = integer_check()
exp = exponent_check()
print(expression(base, exp))
📝 Notes
Many commenters emphasized using a function, as the exercise requires.

The pow() version is simplest.

🎉 Batch 5 Completed (Exercises 13–15)
✅ BATCH 6 — Exercises 16 to 20
⭐ EXERCISE 16 — Palindrome (Function Version)
Question
Write a function to check if a number is a palindrome.

✔️ Canonical Solution
python
def palindrome(num):
    return str(num) == str(num)[::-1]
⭐ Alternative Solutions (from your document)
1. Expanded version
python
def palindrome(num):
    numeral = num
    num_s = str(num)
    num_r = int(num_s[::-1])
    if numeral == num_r:
        print(f"The number {num} is a palindrome")
    else:
        print(f"The number {num} is not a palindrome")
2. Boolean return
python
def is_palindrome(n):
    if str(n) == str(n)[::-1]:
        return True
    else:
        return False
3. Without converting to string
python
def is_palindrome(n):
    original = n
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev == original
📝 Notes
Many commenters insisted on solving without converting to string.

Both approaches are valid.

⭐ EXERCISE 17 — Reverse Number (Math Only)
Question
Reverse a number using a loop.

✔️ Canonical Solution
python
n = 7536
rev = 0

while n > 0:
    rev = rev * 10 + n % 10
    n //= 10

print(rev)
⭐ Alternative Solutions
1. Using string slicing
python
print(str(n)[::-1])
2. Using reversed()
python
print("".join(reversed(str(n))))
3. Printing digits spaced
python
for c in str(n)[::-1]:
    print(c, end=" ")
⭐ EXERCISE 18 — Leap Year Check
Question
Check if a year is a leap year.

✔️ Canonical Solution
python
def is_leap(y):
    return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)
⭐ Alternative Solutions (from your document)
1. Boolean trick
python
def is_leap(y):
    return bool(not y % 4 and y % 100 or not y % 400)
2. Expanded version
python
def is_leap(y):
    if y % 400 == 0:
        return True
    if y % 100 == 0:
        return False
    if y % 4 == 0:
        return True
    return False
⭐ EXERCISE 19 — Reverse Pyramid of Numbers
Question
Print:

Code
5 5 5 5 5
4 4 4 4
3 3 3
2 2
1
✔️ Canonical Solution
python
a = int(input("Enter Range: "))
for i in range(a, 0, -1):
    print((str(i) + " ") * i)
⭐ Alternative Solutions
python
for i in range(5, 0, -1):
    print((str(i) + " ") * i + "\n")
⭐ EXERCISE 20 — Number Pyramid (Generalized)
Question
Print:

Code
1
22
333
4444
...
✔️ Canonical Solution
python
def numPyramid(n):
    for i in range(1, n+1):
        print((str(i) + " ") * i)
⭐ Alternative Solutions
python
for i in range(1, 6):
    print(str(i) * i)
```
