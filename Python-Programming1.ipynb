```python
##PyNative-Basic Python Exercises
Exercise 1. Arithmetic Product and Conditional Logic
Exercise 2. Cumulative Sum of a Range
Exercise 3. String Indexing and Even Slicing
Exercise 4. String Slicing and Substring Removal
Exercise 5. Variable Swapping (The In-Place Method)
Exercise 6. Calculating Factorial with a Loop
Exercise 7. List Manipulation: Add and Remove
Exercise 8. String Reversal
Exercise 9. Vowel Frequency Counter
Exercise 10. Finding Extremes (Min/Max) in a List
Exercise 11. Removing Duplicates from a List
Exercise 12. List Comparison and Boolean Logic
Exercise 13. Filtering Lists with Conditional Logic
Exercise 14. Substring Frequency Analysis
Exercise 15. Nested Loops for Pattern Generation
Exercise 16. Numerical Palindrome Check
Exercise 17. Merging Lists with Parity Filtering
Exercise 18. Integer Digit Extraction and Reversal
Exercise 19. Multi-Tiered Income Tax Calculation
Exercise 20. Nested Loops for Multiplication Tables
Exercise 21. Downward Half-Pyramid Pattern
Exercise 22. Custom Exponentiation Function
Exercise 23. Check Palindrome Number
Exercise 24. Generate Fibonacci Series
Exercise 25. Check Leap Year
Exercise 26. Merging Two Dictionaries
Exercise 27. Finding Common Elements (Intersections)
Exercise 28. Odd/Even List Splitter
Exercise 29. Word Length Analysis
Exercise 30. Word Frequency Counter (The Histogram)
Exercise 31. Print Alternate Prime Numbers
Exercise 32. Dictionary of Squares (Mapping Logic)
Exercise 33. Character Replacer (Data Sanitization)
Exercise 34. Print Reverse Number Pattern
Exercise 35. Digit Detection in Strings
Exercise 36. Capitalize First Letter (Title Case)
Exercise 37. Simple Countdown Timer
Exercise 38. File Creation and Basic I/O
Exercise 39. External File Word Counter
Exercise 40. Introduction to Classes (OOP) ->provide multiple solutions using sql, python separately
See my thinking
You packed a whole mini–foundations course into this list—I love this. I’ll go exercise by exercise, and for each one:

Give a Python solution (sometimes more than one style if it’s interesting).

Give an SQL solution only where it’s natural/meaningful.
For things like file I/O, classes, timers, etc., I’ll note that SQL isn’t a good fit.

I’ll keep each solution minimal and focused so you can turn this into a teaching key.

Exercise 1. Arithmetic product and conditional logic
Task idea: Given two numbers, compute their product and print whether it’s positive, negative, or zero.

Python
python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

product = a * b
print("Product:", product)

if product > 0:
    print("Positive")
elif product < 0:
    print("Negative")
else:
    print("Zero")
Alternative (ternary style):

python
sign = "Positive" if product > 0 else "Negative" if product < 0 else "Zero"
print(sign)
SQL
Assume a table numbers(a INT, b INT).

sql
SELECT
    a,
    b,
    a * b AS product,
    CASE
        WHEN a * b > 0 THEN 'Positive'
        WHEN a * b < 0 THEN 'Negative'
        ELSE 'Zero'
    END AS sign_label
FROM numbers;
Exercise 2. Cumulative sum of a range
Task idea: Sum integers from 1 to n.

Python
python
n = int(input("Enter n: "))

total = 0
for i in range(1, n + 1):
    total += i

print("Sum =", total)
Alternative (formula):

python
total = n * (n + 1) // 2
SQL
Generate numbers 1..n using a numbers table or CTE (example for n = 100):

sql
WITH RECURSIVE nums AS (
    SELECT 1 AS n
    UNION ALL
    SELECT n + 1 FROM nums WHERE n < 100
)
SELECT SUM(n) AS total_sum
FROM nums;
Exercise 3. String indexing and even slicing
Task idea: Given a string, print characters at even indices.

Python
python
s = input("Enter a string: ")

# even indices: 0, 2, 4, ...
even_chars = s[0::2]
print(even_chars)
Explicit loop:

python
result = ""
for i in range(0, len(s), 2):
    result += s[i]
print(result)
SQL
Assume table texts(txt VARCHAR) and DB with SUBSTRING and GENERATE_SERIES (PostgreSQL):

sql
WITH chars AS (
    SELECT
        txt,
        generate_series(1, length(txt)) AS pos
    FROM texts
)
SELECT
    txt,
    string_agg(SUBSTRING(txt FROM pos FOR 1), '' ORDER BY pos) AS even_chars
FROM chars
WHERE (pos - 1) % 2 = 0
GROUP BY txt;
Exercise 4. String slicing and substring removal
Task idea: Remove a given substring from a string.

Python
python
s = input("Enter main string: ")
sub = input("Enter substring to remove: ")

result = s.replace(sub, "")
print(result)
Remove only first occurrence:

python
result = s.replace(sub, "", 1)
SQL
sql
-- Remove all occurrences
SELECT
    txt,
    REPLACE(txt, sub, '') AS removed
FROM my_table;
Where my_table(txt, sub) holds the string and substring.

Exercise 5. Variable swapping (the in-place method)
Python
python
a = int(input("a: "))
b = int(input("b: "))

# Pythonic
a, b = b, a
print(a, b)
Classic temp:

python
temp = a
a = b
b = temp
SQL
SQL doesn’t have variables in the same way, but you can show a conceptual swap in a SELECT:

sql
SELECT
    b AS a_swapped,
    a AS b_swapped
FROM (VALUES (10, 20)) AS t(a, b);
Exercise 6. Calculating factorial with a loop
Python
python
n = int(input("Enter n: "))

fact = 1
for i in range(1, n + 1):
    fact *= i

print(f"Factorial of {n} is {fact}")
Recursive:

python
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)
SQL
Using recursive CTE (PostgreSQL-style):

sql
WITH RECURSIVE fact(n, f) AS (
    SELECT 1, 1
    UNION ALL
    SELECT n + 1, f * (n + 1)
    FROM fact
    WHERE n < 5
)
SELECT f AS factorial_of_5
FROM fact
WHERE n = 5;
Exercise 7. List manipulation: add and remove
Python
python
nums = [1, 2, 3]

nums.append(4)
nums.insert(0, 0)
nums.remove(2)      # remove first occurrence of 2
popped = nums.pop() # remove last element

print(nums, popped)
SQL
Lists map to tables. “Add” = INSERT, “remove” = DELETE.

sql
-- Add
INSERT INTO numbers (value) VALUES (4);

-- Remove
DELETE FROM numbers WHERE value = 2;
Exercise 8. String reversal
Python
python
s = input("Enter string: ")

print(s[::-1])
Loop:

python
rev = ""
for ch in s:
    rev = ch + rev
print(rev)
SQL
sql
WITH RECURSIVE chars AS (
    SELECT 1 AS pos, txt, SUBSTRING(txt, 1, 1) AS ch
    FROM my_table
    UNION ALL
    SELECT pos + 1,
           txt,
           SUBSTRING(txt, pos + 1, 1)
    FROM chars
    WHERE pos < length(txt)
)
SELECT
    txt,
    string_agg(ch, '' ORDER BY pos DESC) AS reversed
FROM chars
GROUP BY txt;
Exercise 9. Vowel frequency counter
Python
python
s = input("Enter string: ").lower()
vowels = "aeiou"
freq = {v: 0 for v in vowels}

for ch in s:
    if ch in vowels:
        freq[ch] += 1

print(freq)
SQL
sql
WITH chars AS (
    SELECT
        txt,
        LOWER(SUBSTRING(txt FROM i FOR 1)) AS ch
    FROM my_table,
         generate_series(1, length(txt)) AS i
)
SELECT
    ch AS vowel,
    COUNT(*) AS freq
FROM chars
WHERE ch IN ('a','e','i','o','u')
GROUP BY ch;
Exercise 10. Finding extremes (min/max) in a list
Python
python
nums = [3, 1, 9, 4]

print("Min:", min(nums))
print("Max:", max(nums))
Manual:

python
mn = mx = nums[0]
for x in nums[1:]:
    if x < mn:
        mn = x
    if x > mx:
        mx = x
print(mn, mx)
SQL
sql
SELECT MIN(value) AS min_val, MAX(value) AS max_val
FROM numbers;
Exercise 11. Removing duplicates from a list
Python
python
nums = [1, 2, 2, 3, 1]

# order-preserving
seen = set()
result = []
for x in nums:
    if x not in seen:
        seen.add(x)
        result.append(x)

print(result)
Set only (order not guaranteed):

python
list(set(nums))
SQL
sql
SELECT DISTINCT value
FROM numbers;
Exercise 12. List comparison and boolean logic
Task idea: Check if two lists are equal (same elements, same order).

Python
python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True/False
As sets (ignoring order):

python
print(set(a) == set(b))
SQL
Assume two tables list_a(pos, val) and list_b(pos, val).

sql
SELECT
    CASE
        WHEN COUNT(*) = 0 THEN 'Equal'
        ELSE 'Not equal'
    END AS comparison
FROM (
    SELECT 'A' AS list_name, pos, val FROM list_a
    UNION ALL
    SELECT 'B', pos, val FROM list_b
) t
GROUP BY pos
HAVING COUNT(DISTINCT val) > 1;
If no rows returned → equal; if rows returned → not equal. For teaching, you can wrap in a CASE around an EXISTS.

Exercise 13. Filtering lists with conditional logic
Python
python
nums = [1, 2, 3, 4, 5, 6]

evens = [x for x in nums if x % 2 == 0]
print(evens)
SQL
sql
SELECT value
FROM numbers
WHERE value % 2 = 0;
Exercise 14. Substring frequency analysis
Python
python
s = "banana bandana"
sub = "ana"

count = 0
start = 0
while True:
    idx = s.find(sub, start)
    if idx == -1:
        break
    count += 1
    start = idx + 1  # allow overlaps
print(count)
SQL
Many DBs don’t have built-in overlapping substring count; one approach (PostgreSQL):

sql
WITH RECURSIVE positions AS (
    SELECT 1 AS pos
    UNION ALL
    SELECT pos + 1 FROM positions WHERE pos < length(txt)
)
SELECT
    txt,
    COUNT(*) AS freq
FROM my_table m
JOIN positions p ON SUBSTRING(m.txt FROM p.pos FOR length(:sub)) = :sub
GROUP BY txt;
Exercise 15. Nested loops for pattern generation
Task idea: Print a right triangle of *.

Python
python
rows = int(input("Rows: "))

for i in range(1, rows + 1):
    line = ""
    for j in range(i):
        line += "*"
    print(line)
SQL
Pattern printing is not SQL’s domain, but you can generate rows:

sql
WITH RECURSIVE nums AS (
    SELECT 1 AS n
    UNION ALL
    SELECT n + 1 FROM nums WHERE n < 5
)
SELECT
    n,
    repeat('*', n) AS pattern
FROM nums;
Exercise 16. Numerical palindrome check
Python
python
n = int(input("Enter number: "))
s = str(n)

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
SQL
sql
SELECT
    n,
    CASE
        WHEN CAST(n AS TEXT) = REVERSE(CAST(n AS TEXT))
        THEN 'Palindrome'
        ELSE 'Not palindrome'
    END AS result
FROM numbers;
Exercise 17. Merging lists with parity filtering
Task idea: Merge two lists and keep only even numbers.

Python
python
a = [1, 2, 3]
b = [4, 5, 6]

merged = a + b
evens = [x for x in merged if x % 2 == 0]
print(evens)
SQL
Assume list_a(value) and list_b(value).

sql
SELECT value
FROM (
    SELECT value FROM list_a
    UNION ALL
    SELECT value FROM list_b
) t
WHERE value % 2 = 0;
Exercise 18. Integer digit extraction and reversal
Python
python
n = int(input("Enter number: "))

digits = []
temp = n
while temp > 0:
    digits.append(temp % 10)
    temp //= 10

print("Digits (reversed order):", digits)
print("Reversed number:", int("".join(map(str, digits))))
String-based:

python
s = str(n)
digits = [int(ch) for ch in s]
rev = int(s[::-1])
SQL
sql
SELECT
    n,
    REVERSE(CAST(n AS TEXT)) AS reversed_text
FROM numbers;
Exercise 19. Multi-tiered income tax calculation
Task idea: Piecewise tax brackets.

Example brackets:

0–10,000 → 0%

10,001–50,000 → 10% on amount above 10,000

50,000 → 10% on 40,000 + 20% on amount above 50,000

Python
python
income = float(input("Income: "))

tax = 0.0
if income <= 10000:
    tax = 0
elif income <= 50000:
    tax = (income - 10000) * 0.10
else:
    tax = (50000 - 10000) * 0.10 + (income - 50000) * 0.20

print("Tax:", tax)
SQL
sql
SELECT
    income,
    CASE
        WHEN income <= 10000 THEN 0
        WHEN income <= 50000 THEN (income - 10000) * 0.10
        ELSE (50000 - 10000) * 0.10 + (income - 50000) * 0.20
    END AS tax
FROM salaries;
Exercise 20. Nested loops for multiplication tables
Python
python
for i in range(1, 11):
    for j in range(1, 11):
        print(f"{i} x {j} = {i*j}")
    print()
SQL
sql
SELECT
    a.n AS i,
    b.n AS j,
    a.n * b.n AS product
FROM generate_series(1, 10) AS a(n)
CROSS JOIN generate_series(1, 10) AS b(n)
ORDER BY i, j;
Exercise 21. Downward half-pyramid pattern
Python
python
rows = int(input("Rows: "))

for i in range(rows, 0, -1):
    print("*" * i)
SQL
sql
WITH RECURSIVE nums AS (
    SELECT 5 AS n
    UNION ALL
    SELECT n - 1 FROM nums WHERE n > 1
)
SELECT
    n,
    repeat('*', n) AS pattern
FROM nums;
Exercise 22. Custom exponentiation function
Python
python
def power(base, exp):
    result = 1
    for _ in range(exp):
        result *= base
    return result

print(power(2, 5))
Using **:

python
base ** exp
SQL
sql
SELECT POWER(2, 5) AS result;
Exercise 23. Check palindrome number
(Same as 16 but explicitly “number”.)

Python
python
n = int(input("Enter number: "))
s = str(n)
print("Palindrome" if s == s[::-1] else "Not palindrome")
SQL
Same pattern as 16.

sql
SELECT
    n,
    CASE
        WHEN CAST(n AS TEXT) = REVERSE(CAST(n AS TEXT))
        THEN 'Palindrome'
        ELSE 'Not palindrome'
    END AS result
FROM numbers;
Exercise 24. Generate Fibonacci series
Python
python
n = int(input("How many terms? "))

a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
SQL
Recursive CTE:

sql
WITH RECURSIVE fib(pos, a, b) AS (
    SELECT 1, 0, 1
    UNION ALL
    SELECT pos + 1, b, a + b
    FROM fib
    WHERE pos < 10
)
SELECT pos, a AS fib_value
FROM fib;
Exercise 25. Check leap year
Python
python
year = int(input("Year: "))

is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print("Leap year" if is_leap else "Not leap year")
SQL
sql
SELECT
    year,
    CASE
        WHEN (year % 4 = 0 AND year % 100 <> 0) OR (year % 400 = 0)
        THEN 'Leap'
        ELSE 'Not leap'
    END AS result
FROM years;
Exercise 26. Merging two dictionaries
Python
python
d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "c": 4}

merged = {**d1, **d2}  # d2 overrides d1
print(merged)
Update:

python
merged = d1.copy()
merged.update(d2)
SQL
Dictionaries map to key/value tables. Merge = UNION or join.

sql
SELECT key, value
FROM dict1
UNION ALL
SELECT key, value
FROM dict2;
If you want “last wins” semantics, use window functions or aggregation.

Exercise 27. Finding common elements (intersections)
Python
python
a = [1, 2, 3, 4]
b = [3, 4, 5]

common = list(set(a) & set(b))
print(common)
SQL
sql
SELECT DISTINCT a.value
FROM list_a a
JOIN list_b b ON a.value = b.value;
Exercise 28. Odd/even list splitter
Python
python
nums = [1, 2, 3, 4, 5, 6]

evens = [x for x in nums if x % 2 == 0]
odds = [x for x in nums if x % 2 != 0]

print("Evens:", evens)
print("Odds:", odds)
SQL
sql
SELECT
    value,
    CASE WHEN value % 2 = 0 THEN 'even' ELSE 'odd' END AS parity
FROM numbers;
Exercise 29. Word length analysis
Task idea: Given a sentence, print each word and its length.

Python
python
sentence = input("Enter sentence: ")

words = sentence.split()
for w in words:
    print(w, len(w))
SQL
Assume sentence column and a word-splitting function (PostgreSQL example using regexp_split_to_table):

sql
SELECT
    word,
    length(word) AS len
FROM (
    SELECT regexp_split_to_table(sentence, '\s+') AS word
    FROM sentences
) t;
Exercise 30. Word frequency counter (the histogram)
Python
python
text = input("Enter text: ")

freq = {}
for word in text.split():
    word = word.lower()
    freq[word] = freq.get(word, 0) + 1

print(freq)
SQL
sql
SELECT
    LOWER(word) AS word,
    COUNT(*) AS freq
FROM (
    SELECT regexp_split_to_table(sentence, '\s+') AS word
    FROM sentences
) t
GROUP BY LOWER(word);
Exercise 31. Print alternate prime numbers
Task idea: Print every second prime in a range (e.g., 2nd, 4th, 6th…).

Python
python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

limit = int(input("Upper limit: "))

primes = [x for x in range(2, limit + 1) if is_prime(x)]
alternate = primes[1::2]  # start from 2nd prime
print(alternate)
SQL
Prime generation in SQL is possible but verbose; for teaching, you can show a simple filter:

sql
WITH nums AS (
    SELECT generate_series(2, 100) AS n
),
primes AS (
    SELECT n
    FROM nums n1
    WHERE NOT EXISTS (
        SELECT 1
        FROM nums n2
        WHERE n2.n < n1.n
          AND n2.n > 1
          AND n1.n % n2.n = 0
    )
),
indexed AS (
    SELECT n, ROW_NUMBER() OVER (ORDER BY n) AS rn
    FROM primes
)
SELECT n
FROM indexed
WHERE rn % 2 = 0;
Exercise 32. Dictionary of squares (mapping logic)
Python
python
n = int(input("Enter n: "))

squares = {i: i*i for i in range(1, n + 1)}
print(squares)
SQL
sql
SELECT
    n AS key,
    n * n AS value
FROM generate_series(1, 10) AS n;
Exercise 33. Character replacer (data sanitization)
Task idea: Replace a given character with another in a string.

Python
python
s = input("Enter string: ")
old = input("Char to replace: ")
new = input("Replacement char: ")

result = s.replace(old, new)
print(result)
SQL
sql
SELECT
    txt,
    REPLACE(txt, :old_char, :new_char) AS sanitized
FROM my_table;
Exercise 34. Print reverse number pattern
Task idea: For n = 5, print:

Code
5 4 3 2 1
4 3 2 1
3 2 1
2 1
1
Python
python
n = int(input("n: "))

for i in range(n, 0, -1):
    line = " ".join(str(x) for x in range(i, 0, -1))
    print(line)
SQL
sql
WITH RECURSIVE rows AS (
    SELECT 5 AS r
    UNION ALL
    SELECT r - 1 FROM rows WHERE r > 1
),
nums AS (
    SELECT r, generate_series(r, 1, -1) AS val
    FROM rows
)
SELECT r, string_agg(val::text, ' ' ORDER BY val DESC) AS pattern
FROM nums
GROUP BY r
ORDER BY r DESC;
Exercise 35. Digit detection in strings
Python
python
s = input("Enter string: ")

has_digit = any(ch.isdigit() for ch in s)
print("Contains digit" if has_digit else "No digits")
SQL
sql
SELECT
    txt,
    CASE
        WHEN txt ~ '[0-9]' THEN 'Contains digit'
        ELSE 'No digits'
    END AS result
FROM my_table;
(Regex syntax may vary by DB.)

Exercise 36. Capitalize first letter (title case)
Python
python
s = input("Enter string: ")

print(s.title())
Only first character of whole string:

python
print(s[0].upper() + s[1:])
SQL
sql
SELECT INITCAP(txt) AS title_cased
FROM my_table;
(Available in PostgreSQL; other DBs may differ.)

Exercise 37. Simple countdown timer
Python
python
import time

n = int(input("Seconds: "))

for i in range(n, 0, -1):
    print(i)
    time.sleep(1)
print("Time's up!")
SQL
Timers are not a natural SQL concept; you generally don’t do sleep-based countdowns in SQL. You can mention that this is application-layer logic, not database logic.

Exercise 38. File creation and basic I/O
Python
python
filename = "example.txt"

with open(filename, "w") as f:
    f.write("Hello, file!\n")

with open(filename, "r") as f:
    content = f.read()

print(content)
SQL
Standard SQL doesn’t handle file creation/reading directly; this is handled by the application or external tools. Some DBs have extensions, but it’s not core SQL.

Exercise 39. External file word counter
Python
python
filename = input("Filename: ")

with open(filename, "r") as f:
    text = f.read()

words = text.split()
print("Word count:", len(words))
SQL
Again, reading external files is not standard SQL. You’d typically load the file into a table first (via ETL) and then run word-count logic similar to Exercise 30.

Exercise 40. Introduction to classes (OOP)
Python
python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hi, I'm {self.name} and I'm {self.age} years old.")

p = Person("Anu", 30)
p.greet()
Alternative with @dataclass:

python
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int

    def greet(self):
        print(f"Hi, I'm {self.name} and I'm {self.age} years old.")
SQL
SQL is declarative and not object-oriented; there’s no direct concept of classes. You can connect the idea conceptually:

A class with fields → a table with columns.

An instance → a row.

Example:

sql
CREATE TABLE person (
    id   SERIAL PRIMARY KEY,
    name VARCHAR(100),
    age  INT
);
But behavior/methods live in application code, not in SQL.
```
