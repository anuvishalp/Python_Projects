```python
Set Operations (Mathematical)
Python sets shine here.

🔵 Union — combine elements
Method 1: |
python
a = {1, 2}
b = {2, 3}
print(a | b)     # {1, 2, 3}
Method 2: union()
python
print(a.union(b))
🔵 Intersection — common elements
Method 1: &
python
print(a & b)     # {2}
Method 2: intersection()
python
print(a.intersection(b))
🔵 Difference — elements in A not in B
Method 1: -
python
print(a - b)     # {1}
Method 2: difference()
python
print(a.difference(b))
🔵 Symmetric Difference — elements NOT common
Method 1: ^
python
print(a ^ b)     # {1, 3}
Method 2: symmetric_difference()
python
print(a.symmetric_difference(b))
🟩 7. Subset / Superset / Disjoint
✔️ Subset
python
a = {1, 2}
b = {1, 2, 3}
print(a.issubset(b))   # True
✔️ Superset
python
print(b.issuperset(a)) # True
✔️ Disjoint
python
x = {1, 2}
y = {3, 4}
print(x.isdisjoint(y)) # True
🟩 8. Frozen Sets (Immutable Sets)
Useful when you need a set as a dictionary key.

python
fs = frozenset([1, 2, 3])
# fs.add(4) → ERROR
🟩 9. Set Comprehensions
python
s = {x*x for x in range(5)}
print(s)   # {0, 1, 4, 9, 16}
🟩 10. Real‑World Scenarios
✔️ Remove duplicates from a list
python
nums = [1, 2, 2, 3, 3, 3]
unique = list(set(nums))
print(unique)
✔️ Fast membership test
python
allowed = {"admin", "manager"}
role = "admin"

if role in allowed:
    print("Access granted")
✔️ Find common students in two classes
python
classA = {"Anu", "Ravi", "John"}
classB = {"Ravi", "Priya"}

print(classA & classB)   # {'Ravi'}
✔️ Find items only in one list
python
a = {1, 2, 3}
b = {3, 4, 5}
print(a - b)   # {1, 2}
🟦 Summary Table
Operation	Symbol	Method
Union	`	`	union()
Intersection	&	intersection()
Difference	-	difference()
Symmetric Difference	^	symmetric_difference()
Subset	—	issubset()
Superset	—	issuperset()
Disjoint	—	isdisjoint()
```

```python
🧩 Python json Module — Complete Cheatsheet
The json module handles serialization (Python → JSON) and deserialization (JSON → Python).

📌 1. Core Functions Overview
Function	Direction	Purpose
json.dumps()===========>Python → JSON (string)===========>	Convert Python object to JSON string
json.dump()	===========>Python → JSON (file)	===========>Write JSON to a file
json.loads()===========>	JSON (string) → Python===========>	Parse JSON string into Python object
json.load()	===========>JSON (file) → Python	===========>Read JSON from file and convert to Python
json.JSONEncoder===========>	Python → JSON	Custom encoding
json.JSONDecoder===========>	JSON → Python	Custom decoding

🧠 2. Supported Python → JSON Type Mapping
Python Type===========>	JSON Equivalent
dict===========>	object
list, tuple===========>	array
str	===========>string
int, float	===========>number
True===========>	true
False===========>	false
None===========>	null

🚀 3. json.dumps() — Convert Python Object to JSON String

✔ Basic Usage

      import json

      data = {"name": "Anu", "age": 30}
      json_str = json.dumps(data)
      print(json_str)

✔ Pretty Printing
      json.dumps(data, indent=4)
✔ Sorting Keys
      json.dumps(data, sort_keys=True)
✔ Handling Non‑ASCII Characters

        json.dumps({"city": "Hyderabad"}, ensure_ascii=False)

✔ Custom Separators (compact JSON)
        json.dumps(data, separators=(",", ":"))
✔ Handling Non‑Serializable Objects (default handler)

        import datetime

        def encode_default(obj):
            if isinstance(obj, datetime.datetime):
                return obj.isoformat()
            raise TypeError("Type not serializable")

        json.dumps({"now": datetime.datetime.now()}, default=encode_default)

📁 4. json.dump() — Write JSON to a File
✔ Basic Example

      with open("data.json", "w") as f:
          json.dump(data, f)
✔ Pretty Print to File
      json.dump(data, f, indent=4)
✔ Custom Encoding + File Writing
      json.dump(data, f, ensure_ascii=False)

🔍 5. json.loads() — Parse JSON String to Python
✔ Basic Example
      json_str = '{"name": "Anu", "age": 30}'
      data = json.loads(json_str)
✔ Handling Invalid JSON (try/except)
        try:
            data = json.loads("invalid json")
        except json.JSONDecodeError as e:
            print("Error:", e)
✔ Parse JSON with numbers as decimals
      from decimal import Decimal
      json.loads('{"value": 12.34}', parse_float=Decimal)
✔ Parse integers as custom type
      json.loads('{"value": 10}', parse_int=lambda x: int(x) * 10)
📂 6. json.load() — Read JSON from File

✔ Basic Example
      with open("data.json") as f:
          data = json.load(f)

✔ With Custom Number Parsing
        json.load(f, parse_float=Decimal)

🧬 7. Custom Encoding with JSONEncoder
✔ Custom Encoder Class
                  import json
                  import datetime

                  class MyEncoder(json.JSONEncoder):
                      def default(self, obj):
                          if isinstance(obj, datetime.datetime):
                              return obj.isoformat()
                          return super().default(obj)

                  data = {"now": datetime.datetime.now()}
                  print(json.dumps(data, cls=MyEncoder))

🧬 8. Custom Decoding with JSONDecoder

✔ Convert ISO date strings back to datetime

        import json
        from datetime import datetime

        def decode_datetime(dct):
            for key, value in dct.items():
                if isinstance(value, str) and value.endswith("Z"):
                    try:
                        dct[key] = datetime.fromisoformat(value.replace("Z", "+00:00"))
                    except:
                        pass
            return dct

        json_str = '{"timestamp": "2024-01-01T10:00:00Z"}'
        data = json.loads(json_str, object_hook=decode_datetime)

🧩 9. Advanced Parameters Cheatsheet

Parameter===========>	Used In===========>	Purpose
indent===========>	dumps, dump	===========>Pretty print
sort_keys	===========>dumps, dump	===========>Sort keys alphabetically
ensure_ascii===========>	dumps, dump	===========>Allow Unicode
separators===========>	dumps, dump	===========>Control separators
default===========>	dumps, dump	===========>Handle unsupported types
cls	===========>dumps, dump	===========>Custom encoder
object_hook===========>	loads, load	===========>Custom dict decoding
parse_float===========>	loads, load	===========>Convert floats
parse_int===========>	loads, load	===========>Convert integers
parse_constant===========>	loads, load	===========>Handle NaN, Infinity
=================================================
🧪 10. Real‑World Scenarios

🌍 Scenario 1: Convert API Response to Python

      response = '{"status": "ok", "items": [1, 2, 3]}'
      data = json.loads(response)

💾 Scenario 2: Save Config File

      config = {"theme": "dark", "version": 1.0}

      with open("config.json", "w") as f:
          json.dump(config, f, indent=4)

🔄 Scenario 3: Read Config File

        with open("config.json") as f:
            config = json.load(f)

🕒 Scenario 4: Serialize datetime objects

        json.dumps({"created": datetime.datetime.now()}, default=str)

🧱 Scenario 5: Convert Python Object → JSON → Python

            data = {"a": 1, "b": [1, 2, 3]}
            json_str = json.dumps(data)
            data2 = json.loads(json_str)
🧨 Scenario 6: Handling NaN, Infinity

            json.dumps(float('nan'))  # "NaN"
            json.dumps(float('inf'))  # "Infinity"

🧰 Scenario 7: Custom Object Serialization

            class User:
                def __init__(self, name):
                    self.name = name

            def encode_user(obj):
                if isinstance(obj, User):
                    return {"User": obj.name}
                raise TypeError

            json.dumps(User("Anu"), default=encode_user)

🏁 Complete Summary Table
Task===========>	Function	===========>Example
Python → JSON string===========>	dumps()	===========>json.dumps(obj)
Python → JSON file===========>	dump()===========>	json.dump(obj, f)
JSON string → Python===========>	loads()	===========>json.loads(s)
JSON file → Python===========>	load()===========>	json.load(f)
Custom encoding===========>	JSONEncoder===========>	cls=MyEncoder
Custom decoding	===========>JSONDecoder===========>	object_hook=func
```
