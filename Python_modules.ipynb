```python
Set Operations (Mathematical)
Python sets shine here.

🔵 Union — combine elements
Method 1: |
python
a = {1, 2}
b = {2, 3}
print(a | b)     # {1, 2, 3}
Method 2: union()
python
print(a.union(b))
🔵 Intersection — common elements
Method 1: &
python
print(a & b)     # {2}
Method 2: intersection()
python
print(a.intersection(b))
🔵 Difference — elements in A not in B
Method 1: -
python
print(a - b)     # {1}
Method 2: difference()
python
print(a.difference(b))
🔵 Symmetric Difference — elements NOT common
Method 1: ^
python
print(a ^ b)     # {1, 3}
Method 2: symmetric_difference()
python
print(a.symmetric_difference(b))
🟩 7. Subset / Superset / Disjoint
✔️ Subset
python
a = {1, 2}
b = {1, 2, 3}
print(a.issubset(b))   # True
✔️ Superset
python
print(b.issuperset(a)) # True
✔️ Disjoint
python
x = {1, 2}
y = {3, 4}
print(x.isdisjoint(y)) # True
🟩 8. Frozen Sets (Immutable Sets)
Useful when you need a set as a dictionary key.

python
fs = frozenset([1, 2, 3])
# fs.add(4) → ERROR
🟩 9. Set Comprehensions
python
s = {x*x for x in range(5)}
print(s)   # {0, 1, 4, 9, 16}
🟩 10. Real‑World Scenarios
✔️ Remove duplicates from a list
python
nums = [1, 2, 2, 3, 3, 3]
unique = list(set(nums))
print(unique)
✔️ Fast membership test
python
allowed = {"admin", "manager"}
role = "admin"

if role in allowed:
    print("Access granted")
✔️ Find common students in two classes
python
classA = {"Anu", "Ravi", "John"}
classB = {"Ravi", "Priya"}

print(classA & classB)   # {'Ravi'}
✔️ Find items only in one list
python
a = {1, 2, 3}
b = {3, 4, 5}
print(a - b)   # {1, 2}
🟦 Summary Table
Operation	Symbol	Method
Union	`	`	union()
Intersection	&	intersection()
Difference	-	difference()
Symmetric Difference	^	symmetric_difference()
Subset	—	issubset()
Superset	—	issuperset()
Disjoint	—	isdisjoint()
```
