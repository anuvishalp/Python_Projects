```python
##Step 13.b. Create a lambda function and change the type of item price
dollarizer = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(dollarizer)
```

```python
##Step 13.c. Check the item price type
chipo.item_price.dtype
```

```python
##Step 14. How much was the revenue for the period in the dataset?
revenue = (chipo['quantity']* chipo['item_price']).sum()

print('Revenue was: $' + str(np.round(revenue,2)))
```

```python
##Step 15. How many orders were made in the period?
orders = chipo.order_id.value_counts().count()
orders
```

```python
##Step 16. What is the average revenue amount per order?
## Solution 1

chipo['revenue'] = chipo['quantity'] * chipo['item_price']
order_grouped = chipo.groupby(by=['order_id']).sum()
order_grouped.mean()['revenue']
```

```python
order_grouped = chipo.groupby('order_id').sum(numeric_only=True)
order_grouped['revenue'].mean()
```

```python
##⭐ Fix 2 — Drop non‑numeric columns before grouping
numeric_cols = ['quantity', 'item_price', 'revenue']
order_grouped = chipo[numeric_cols + ['order_id']].groupby('order_id').sum()
order_grouped['revenue'].mean()
```

```python
##⭐ Fix 3 — Use .agg() to explicitly control what gets aggregated
order_grouped = chipo.groupby('order_id').agg({
    'quantity': 'sum',
    'item_price': 'sum',
    'revenue': 'sum'
})

order_grouped['revenue'].mean()
```
