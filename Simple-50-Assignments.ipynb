```python
## 41-42 questions of python

⭐ 41 – School Menu System (with all instructions applied)

✅ Requirements Recap
        Use a list to store student names
        Use append() to admit
        Use remove() to remove
        Use for loop to display
        Wrap everything in a while True loop
        Use break to exit

⭐ Final Code (Perfect for Assignment Submission)

          students = []   # list to store student names

          while True:
              print("\n----- School Menu -----")
              print("1 - Admit student")
              print("2 - Remove student")
              print("3 - View all students")
              print("4 - Exit")

              choice = input("Enter your choice: ")

              # 1 - Admit student
              if choice == "1":
                  name = input("Enter student name to admit: ")
                  students.append(name)   # list built-in method
                  print(name, "has been admitted.")

              # 2 - Remove student
              elif choice == "2":
                  name = input("Enter student name to remove: ")

                  if name in students:    # check if student exists
                      students.remove(name)
                      print(name, "has been removed.")
                  else:
                      print("Student not found in school.")

              # 3 - View all students
              elif choice == "3":
                  print("\n--- List of Students ---")
                  for s in students:      # iterate list
                      print(s)

              # 4 - Exit
              elif choice == "4":
                  print("Exiting system...")
                  break                   # stop while loop

              else:
                  print("Invalid choice! Please try again.")
This version matches every instruction your teacher expects.
-----------------------------------------------------------

⭐ 42 – Hospital Management System

🎯 Goals

        Store name + age of each patient
        Maximum capacity = 40 patients
        Reject patients older than 60
        If user enters more than 40 patients → show message and stop
        Use a list of lists → [ [name, age], [name, age], ... ]

        After storing, show how many more patients can be accommodated

⭐ Final Code (Clear, Simple, Assignment‑Perfect)

        hospital = []   # major list to store all patients

        limit = 40
        num = int(input("Enter number of patients: "))

        # If user enters more than 40, stop immediately
        if num > limit:
            print("We cannot accommodate more than 40 patients.")
        else:
            for i in range(num):
                print("\nEnter data for patient", i+1)
                name = input("Name: ")
                age = int(input("Age: "))

                # Reject old age patients
                if age > 60:
                    print("We do not admit patients older than 60. Skipping this patient.")
                    continue

                # Store patient data
                patient = [name, age]
                hospital.append(patient)

            # Show remaining capacity
            remaining = limit - len(hospital)
            print("\nTotal admitted patients:", len(hospital))
            print("We can still accommodate", remaining, "more patients.")

            # Display all admitted patients
            print("\n--- List of Admitted Patients ---")
            for p in hospital:
                print("Name:", p[0], "| Age:", p[1])


  ⭐ Why This Code Is Perfect

Uses list of lists exactly as required
Uses conditional logic to reject age > 60
Uses continue to skip invalid patients
Uses for loop to collect data
Uses capacity logic correctly
Clean, readable, beginner‑friendly
```

```python
### 43 to 50 questions of Python

⭐ 43 – Employee Salaries Management System
Requirements Recap
    Use a while loop
    Stop when employee ID = 0
    Pay overtime (1.5×) for hours > 40
    Deduct 3.625% tax

Track:

    total salary of all employees
    total number of employees

✅ Final Code

          total_salary = 0
          total_employees = 0

          while True:
              emp_id = int(input("Enter employee ID (0 to stop): "))

              if emp_id == 0:
                  break

              hourly_wage = float(input("Enter hourly wage: "))
              hours_worked = float(input("Enter hours worked: "))

              # Calculate gross salary
              if hours_worked > 40:
                  overtime = hours_worked - 40
                  gross = (40 * hourly_wage) + (overtime * hourly_wage * 1.5)
              else:
                  gross = hours_worked * hourly_wage

              # Deduct tax
              tax = gross * 0.03625
              net_salary = gross - tax

              print("Net salary of employee:", net_salary)

              total_salary += net_salary
              total_employees += 1

          print("\nTotal employees:", total_employees)
          print("Total salary paid:", total_salary)
-----------------------------------------------------------

⭐ 44 – Regular Budget Management System
Requirements

Function budget(available)
Keep asking expenses until user enters 0

Print:

total expenses
remaining balance
surplus / deficit / balanced

✅ Final Code

        def budget(available):
            total_expenses = 0

            while True:
                expense = float(input("Enter expense (0 to stop): "))
                if expense == 0:
                    break
                total_expenses += expense

            print("\nTotal monthly budget:", available)
            print("Total expenses:", total_expenses)

            remaining = available - total_expenses
            print("Remaining balance:", remaining)

            if remaining > 0:
                print("Status: Surplus")
            elif remaining < 0:
                print("Status: Deficit")
            else:
                print("Status: Balanced")

        # Example call
        budget(50000)
-----------------------------------------------------------

⭐ 45 – Restaurant Menu System
Requirements
Use list indexing to get price

Ask for quantity

If user chooses shake → ask flavor

✅ Final Code
python
def res_menu():
    print('=================MENU===============')
    print('1 - Burger beef - 600')
    print('2 - Tortilla Wraps - 800')
    print('3 - Chai Paratha - 200')
    print('4 - Chai - 150')
    print('5 - Shake (Mango, banana, date, chiku) - 600')

    menu = [600, 800, 200, 150, 600]

    choice = int(input("\nEnter your option: "))

    if choice == 5:
        flavor = input("Enter shake flavor: ")

    qty = int(input("Enter quantity: "))

    price = menu[choice - 1]
    total = price * qty

    print("Total bill:", total)

res_menu()
⭐ 46 – Advanced Restaurant Menu System
Requirements
Allow multiple orders

Stop when user enters 0

Show total bill at end

✅ Final Code
python
def restaurant():
    menu = [600, 800, 200, 150, 600]
    total_bill = 0

    while True:
        print("\n=================MENU===============")
        print("1 - Burger beef - 600")
        print("2 - Tortilla Wraps - 800")
        print("3 - Chai Paratha - 200")
        print("4 - Chai - 150")
        print("5 - Shake - 600")
        print("0 - Finish order")

        choice = int(input("Enter your option: "))

        if choice == 0:
            break

        if choice == 5:
            flavor = input("Enter shake flavor: ")

        qty = int(input("Enter quantity: "))
        total_bill += menu[choice - 1] * qty

    print("\nYour total bill is:", total_bill)

restaurant()
⭐ 47 – Food Cost Management
Requirements
Ask breakfast, lunch, supper cost for each day

Ask if user wants to enter another day

Track:

total breakfast

total lunch

total supper

total meals

✅ Final Code
python
def food_cost():
    total_breakfast = 0
    total_lunch = 0
    total_supper = 0

    while True:
        b = float(input("Breakfast cost: "))
        l = float(input("Lunch cost: "))
        s = float(input("Supper cost: "))

        total_breakfast += b
        total_lunch += l
        total_supper += s

        again = input("Enter data for another day? (y/n): ")
        if again.lower() != 'y':
            break

    print("\nTotal breakfast cost:", total_breakfast)
    print("Total lunch cost:", total_lunch)
    print("Total supper cost:", total_supper)
    print("Total cost of all meals:", total_breakfast + total_lunch + total_supper)

food_cost()
⭐ 48 – Simple Visualizations (Sales Bar Chart)
Requirements
5 shops

Each 100 rupees = 1 *

Print bar chart

✅ Final Code
python
sales = []

for i in range(1, 6):
    amount = int(input(f"Enter today's sales for store {i}: "))
    sales.append(amount)

print("\nSALES BAR CHART")

for i in range(5):
    stars = sales[i] // 100
    print(f"Store {i+1}: {'*' * stars}")
⭐ 49 – Student Marks Bar Chart
Requirements
Function takes list of marks

Print roll number + marks

Print bar chart (1 star per mark)

✅ Final Code
python
def marks_chart(marks):
    print("Roll No | Marks")
    for i in range(len(marks)):
        print(i+1, "->", marks[i])

    print("\nBar Chart:")
    for i in range(len(marks)):
        print("Roll", i+1, ":", "*" * marks[i])

marks_chart([23, 43, 50, 12])
⭐ 50 – Rock, Paper, Scissors Game
Requirements
Use list: ['rock', 'paper', 'scissor']

Computer chooses randomly

User chooses

Decide winner

Put everything in a function

✅ Final Code
python
import random

def rps():
    options = ['rock', 'paper', 'scissor']

    computer = random.choice(options)
    user = input("Enter rock, paper, or scissor: ").lower()

    print("Computer chose:", computer)

    if user == computer:
        print("It's a tie!")
    elif (user == "rock" and computer == "scissor") or \
         (user == "scissor" and computer == "paper") or \
         (user == "paper" and computer == "rock"):
        print("You win!")
    else:
        print("Computer wins!")

# Play again and again
while True:
    rps()
    again = input("Play again? (y/n): ")
    if again.lower() != 'y':
        break
```
